```python
import pandas as pd
```

```python
#q1
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
```

```python
import requests
import io

#Read all the tables in website
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response = requests.get(url, headers=headers)
tables = pd.read_html(io.StringIO(response.text))
```

```python
#the first table contains the list of companies
sp500_table = tables[1]
```

```python
#extract the tickers
tickers = sp500_table['Symbol'].tolist()
```

```python
#clean tickers for example removing stickers
tickers = [t.replace('.', '-') for t in tickers]
sp500_table
```

```python
import yfinance as yf

start_date = "2022-01-01"
end_date = "2025-01-01"

#download price data
data = yf.download(tickers, start=start_date, end=end_date, group_by='tickers')
```

```python
sp500_df=pd.DataFrame(data)
sp500_df
```

```python
data.isna().sum()
```

```python
#2a
# Daily returns for all stocks
# Select only 'Close' prices for calculating returns
close_prices = data.xs('Close', level='Price', axis=1)

# Drop columns (tickers) that have any NaN values in their close prices
close_prices = close_prices.dropna(axis=1)

df_returns = close_prices.pct_change().dropna()
df_returns.head()
```

```python
#2b
spy = yf.download("SPY", start=start_date, end=end_date)['Close']
spy_returns = spy.pct_change().dropna()
```

```python
import numpy as np

betas = {}

for ticker in df_returns.columns:

    # Align stock and index returns
    stock_ret = df_returns[ticker].dropna()
    # Ensure market_ret is a Series for proper alignment and correlation calculation
    market_ret = spy_returns['SPY']
    aligned_index, aligned_stock = market_ret.align(stock_ret, join='inner')

    # Compute correlation
    corr = np.corrcoef(aligned_stock, aligned_index)[0, 1]

    # Standard deviations
    stock_std = aligned_stock.std()
    index_std = aligned_index.std()

    # Beta formula
    beta = corr * (stock_std / index_std)
    betas[ticker] = beta
```

```python
df_beta = pd.DataFrame.from_dict(betas, orient='index', columns=['Beta'])
df_beta
```

```python
#2C
daily_vol = df_returns.std()
annual_vol = daily_vol * np.sqrt(252)

df_volatility = annual_vol.to_frame(name='Annual_Volatility')
df_volatility
```

```python
#METRICS FOR CLUSTERING
df_metrics = pd.concat([df_returns.mean().to_frame('Avg_Daily_Return'),
                        df_beta,
                        df_volatility],
                        axis=1)

df_metrics.head()
```

```python
#q3
#dendrogram
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
import numpy as np

# Beta values (1D)
beta_values = df_beta['Beta'].values.reshape(-1, 1)

# Perform hierarchical clustering
linked = sch.linkage(beta_values, method='ward')

# Plot dendrogram
plt.figure(figsize=(14, 6))
dend = sch.dendrogram(linked, labels=df_beta.index, leaf_rotation=90, leaf_font_size=8)

# Annotate the cut for k=2
# Find the height where we cut the dendrogram to get 2 clusters
# In 1D Beta, the largest vertical gap corresponds to ~max distance
max_distance = np.max(linked[:, 2])
cut_height = max_distance * 0.5  # approx middle of last merge

plt.axhline(y=cut_height, c='red', linestyle='--', label='Cut for k=2')
plt.title("Dendrogram for Stock Beta Clustering (k=2)")
plt.xlabel("Stocks")
plt.ylabel("Euclidean Distance")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import fcluster

best_k = None
best_score = -1

for k in range(2, 11):
    labels = fcluster(linked, k, criterion='maxclust')
    score = silhouette_score(beta_values, labels)
    print(f"K={k}, Silhouette={score:.4f}")

    if score > best_score:
        best_score = score
        best_k = k

k_silhouette = best_k
print("\nK from Silhouette:", k_silhouette)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.metrics import silhouette_score

# ---------------------------
# 1. Prepare data and linkage
# ---------------------------
# assumes beta_df exists
X = df_beta[['Beta']].values

# if Z already exists you can skip this line
Z = linkage(X, method='ward')

# -----------------------------------
# 2. K from dendrogram (largest jump)
# -----------------------------------
distances = Z[:, 2]
gaps = np.diff(distances)
idx = np.argmax(gaps)
n = X.shape[0]
k_dendrogram = n - (idx + 1)
print("K from Dendrogram:", k_dendrogram)

# ------------------------------------
# 3. K from silhouette (best K in 2â€“10)
# ------------------------------------
k_values = range(2, 11)
sil_scores = []

best_k = None
best_score = -1

for k in k_values:
    labels = fcluster(Z, k, criterion='maxclust')
    score = silhouette_score(X, labels)
    sil_scores.append(score)
    if score > best_score:
        best_score = score
        best_k = k

k_silhouette = best_k
print("K from Silhouette:", k_silhouette)

# -------------------------------
# 4. Plot comparison on a graph
# -------------------------------
plt.figure(figsize=(10, 5))

# silhouette curve
plt.plot(list(k_values), sil_scores, marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Silhouette score")
plt.title("Silhouette scores for different K\nwith Dendrogram K vs Silhouette K")

# vertical line for dendrogram K
plt.axvline(x=k_dendrogram, color='red', linestyle='--', label=f"Dendrogram K = {k_dendrogram}")

# vertical line for silhouette K
plt.axvline(x=k_silhouette, color='green', linestyle='--', label=f"Silhouette K = {k_silhouette}")

plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import fcluster

# Ensure df_metrics is defined (copied for robustness, assuming dependencies like df_returns, df_beta, df_volatility are available)
# METRICS FOR CLUSTERING
df_metrics = pd.concat([df_returns.mean().to_frame('Avg_Daily_Return'),
                        df_beta,
                        df_volatility],
                        axis=1)

# Prepare data for clustering (using 'Beta' values as X, consistent with previous clustering)
X = df_beta['Beta'].values.reshape(-1, 1)

# Assuming 'linked' (or 'Z') from hierarchical clustering is available from previous cells.
# If not, it would need to be re-calculated here: linked = sch.linkage(X, method='ward')

# Recalculate silhouette scores and K_range for this specific cell's execution
K_range = range(2, 11)  # Define K_range
sil_scores = []

for k in K_range:
    labels = fcluster(linked, k, criterion='maxclust') # Use 'linked' from previous cells
    score = silhouette_score(X, labels)
    sil_scores.append(score)

# Choose best k (highest silhouette score)
best_k = K_range[sil_scores.index(max(sil_scores))]
print("Best k according to silhouette score:", best_k)

# Fit Agglomerative Clustering with best_k
agg_cluster = AgglomerativeClustering(n_clusters=best_k, linkage='ward')
df_metrics['Agg_Cluster'] = agg_cluster.fit_predict(X)

# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_metrics.index, y=df_metrics['Beta'], hue=df_metrics['Agg_Cluster'], palette='Set2', s=100)
plt.xticks(rotation=90)
plt.title(f"Agglomerative Clustering of Stocks (k={best_k})")
plt.xlabel("Stock")
plt.ylabel("Beta")
plt.legend(title="Cluster")
plt.show()

# Cluster profile
agg_profile = df_metrics.groupby('Agg_Cluster')['Beta'].describe()
print("Agglomerative Clustering Profile:\n", agg_profile)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# METRICS FOR CLUSTERING (copied from cell wzbdyHapvL80)
df_metrics = pd.concat([df_returns.mean().to_frame('Avg_Daily_Return'),
                        df_beta,
                        df_volatility],
                        axis=1)

# Prepare the data for clustering (Beta + Annual Volatility)
X_kmeans = df_metrics[['Beta', 'Annual_Volatility']].values

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_kmeans)

# Choose best k based on elbow or silhouette score
# For example, using the silhouette score result from the previous cell:
best_k_kmeans = 3 # Let's assume K=3 from the silhouette analysis for demonstration

# Fit K-Means with the chosen k
kmeans = KMeans(n_clusters=best_k_kmeans, random_state=42)
df_metrics['KMeans_Cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Beta',
    y='Annual_Volatility',
    hue='KMeans_Cluster',
    palette='Set1',
    data=df_metrics,
    s=100
)
plt.title(f"K-Means Clustering (k={best_k_kmeans})")
plt.xlabel("Beta")
plt.ylabel("Annual Volatility")
plt.legend(title="Cluster")
plt.show()

# Cluster profile
kmeans_profile = df_metrics.groupby('KMeans_Cluster')[['Beta', 'Annual_Volatility']].describe()
print("K-Means Clustering Profile:\n", kmeans_profile)
```
